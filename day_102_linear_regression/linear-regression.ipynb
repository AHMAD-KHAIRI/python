**Linear Regression**

Example 1:
Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```python
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
random_noise = np.random.normal(0, 0.1, 1000)
purchaseAmount = 100 - (pageSpeeds + random_noise) * 3

scatter(pageSpeeds, purchaseAmount)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

The linear equation of a straight line in the Cartesian coordinate system is given by: 
    
    y = mx + b, 
    
    where 'm' is the slope and 'b' is the intercept

In the context of linear regression, where you're trying to fit a line to a set of data points, the equation becomes:
    
    y = slope * x + intercept

```python
print(f"The value of slope 'm' is {slope}")
# slope variable will store the slope of the regression line. 
# In the context of linear regression, the slope represents the change in the dependent variable (in this case, purchaseAmount) for a one-unit change in the independent variable (in this case, pageSpeeds).
```

```python
print(f"The value of intercept 'b' is {intercept}")
# intercept will store the y-intercept of the regression line. 
# The y-intercept is the value of the dependent variable (purchaseAmount) when the independent variable (pageSpeeds) is zero.
```

```python
print(f"The value of r_value is {r_value}")
# r_value stores the correlation coefficient
# correlation coefficient measures the strength and direction of the linear relationship between the two variables. 
# It ranges from -1 to 1
# -1 indicating a perfect negative correlation, 1 indicating a perfect positive correlation, and 0 indicating no correlation.
```

```python
print(f"The value of p_value is {p_value}")
# This variable will store the p-value associated with the null hypothesis that there is no correlation between the variables. 
# A small p-value (typically less than 0.05) suggests that there is enough evidence to reject the null hypothesis.
```

```python
print(f"The value of std_err is {std_err}")
# This variable will store the standard error of the estimated slope. 
# The standard error provides a measure of the variability of the estimated slope.
```

Now, calculate our R-squared value:

```python
print(f"R-squared value of {r_value} is {r_value ** 2}")
```

Let's plot!

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept # this is the linear equation: y = slope * x + intercept

fitLine = predict(pageSpeeds) # fitline is the y value

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, color='red')
plt.show()
```

Example 2: Let's create a hypothetical example related to industrial automation and robotics. In this case, we'll simulate data related to the time it takes for a robot to complete a task (cycle time) based on the number of components to be assembled.

```python
# Section 1: Generate fabricated data
import numpy as np
from pylab import *


np.random.seed(42)
num_components = np.random.randint(10, 50, 100)
noise = np.random.normal(0, 5, 100)
cycle_time = 2 * num_components + 10 + noise

plt.figure(dpi=200)
plt.scatter(num_components, cycle_time, label='Actual Data')
plt.xlabel('Number of Components')
plt.ylabel('Cycle Time')
plt.title('Generated Data for Linear Regression')
plt.legend()
plt.show()
```

```python
num_components
```

```python
noise
```

```python
cycle_time
```

```python
# Section 2: Calculate Linear Regression
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(num_components, cycle_time)
```

```python
# dependent variable: cycle time, independent variable: num_components
print(f"The slope 'm' value is {slope}")
print(f"The intercept 'b' value is {intercept}")
print(f"The r_value is {r_value}")
print(f"The p_value is {p_value}")
print(f"The value of std_err is {std_err}")
```

```python
# Section 3: Calculate R-squared
r_squared = r_value**2
print(f'R-squared: {r_squared} ({r_squared * 100:.4}%)')
```

R-squared = 0: Indicates that the independent variable does not explain any of the variability in the dependent variable.

R-squared = 1: Indicates that the independent variable explains all of the variability in the dependent variable.

```python
# Section 4: Plot Generated Data and Linear Regression
import matplotlib.pyplot as plt

def predict_cycle_time(x):
    return slope * x + intercept # y = mx + b

fit_line = predict_cycle_time(num_components)

plt.figure(dpi=200)
plt.scatter(num_components, cycle_time, label='Actual Data')
plt.plot(num_components, fit_line, color='red', label='Regression Line')
plt.xlabel('Number of Components')
plt.ylabel('Cycle Time')
plt.title(f'Linear Regression in Industrial Automation (R = {r_squared:.4})')
plt.legend()
plt.show()
```

Instead of using <code>scipy - stats - linregress</code>, let's try to use the <code>scikit-learn</code> library:

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

np.random.seed(42)
num_components = np.random.randint(10, 50, 100)
noise = np.random.normal(0, 5, 100)
cycle_time = 2 * num_components + 10 + noise

# Reshape the input data for scikit-learn
X = num_components.reshape(-1, 1)

# Create and fit a linear regression model
model = LinearRegression()
model.fit(X, cycle_time)

# Obtain predictions from the model
predictions = model.predict(X)

# Calculate R^2 score
r_squared = r2_score(cycle_time, predictions)

# Plot the actual data and the regression line
plt.figure(dpi=200)
plt.scatter(num_components, cycle_time, label='Actual Data')
plt.plot(num_components, predictions, color='red', label='Regression Line')
plt.xlabel('Number of Components')
plt.ylabel('Cycle Time')
plt.title(f'Linear Regression in Industrial Automation (R = {r_squared:.4f})')
plt.legend()
plt.show()
```
