```python
# convert the temperature unit from Kelvin (K) to Celsius (C) on columns "Air temperature [K]" and "Process temperature [K]" and replace the old columns with new columns
# here is the formula: C = K - 273.15
df["Air temperature [C]"] = df["Air temperature [K]"] - 273.15
df["Process temperature [C]"] = df["Process temperature [K]"] - 273.15
```

```python
df = df.drop("Air temperature [K]", axis=1)
```

```python
df = df.drop("Process temperature [K]", axis=1)
```

```python
# Split the dataset into features (X) and target (y)
X = df.drop("Target", axis=1)
y = df["Target"]
```

```python
# split the dataset into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Type", 
                        "Failure Type"]

one_hot = OneHotEncoder(sparse_output=False)
transformer = ColumnTransformer([("one_hot",
                                one_hot,
                                categorical_features)],
                                remainder="passthrough")

transformed_X_train = transformer.fit_transform(X_train)
transformed_X_test = transformer.fit_transform(X_test)
```

```python
pd.DataFrame(transformed_X_train)
```
